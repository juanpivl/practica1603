```python
from sklearn.datasets import load_iris
```

```python
data= load_iris()
print(data)
```

```python
print('Informacion en el dataset:')
print(data.keys())
print()
```

```python
print('Caracteristicas del dataset')
print(data.DESCR)
```

```python
X = data.data
Y= data.target
```

SVW

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20,random_state=20)
```

```python
from sklearn import svm
from sklearn.svm import SVC
```

```python
clasificador = SVC(kernel ='linear',C =0.025)
clasificador.fit(X_train,Y_train)
print(clasificador.score(X_test,Y_test))
```

Nearest Neighbors

```python
from sklearn import neighbors, datasets, preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python

scaler = preprocessing.StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
knn = neighbors.KNeighborsClassifier(n_neighbors=4)
knn.fit(X_train, Y_train)
Y_pred = knn.predict(X_test)
```

```python
print(accuracy_score(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
print(confusion_matrix(Y_test, Y_pred))
```

Naive Vayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_train,Y_train).predict(X_test)
```

```python
print('Accuracy',gnb.score(X_test,Y_test))
```

Decision tress


```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier(criterion='entropy',max_depth=4,splitter='best')
tree.fit(X_train,Y_train)
```

```python
print('Acuracy',tree.score(X_test,Y_test))
```
